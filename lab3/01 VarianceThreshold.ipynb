```python
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import VarianceThreshold
import numpy as np
import pandas as pd

# Załaduj dane
data = load_breast_cancer()
X, y = data.data, data.target
feature_names = data.feature_names

# Zobacz dane jako DataFrame
df = pd.DataFrame(X, columns=feature_names)
print("Oryginalny kształt danych:", df.shape)

# Oblicz wariancję dla każdej cechy
variances = df.var()
print("\nWariancje cech:\n", variances)

# Ustaw próg – np. usuń cechy o wariancji < 0.01
selector = VarianceThreshold(threshold=0.01)
X_reduced = selector.fit_transform(X)

# Zobacz nowe cechy
selected_mask = selector.get_support()
selected_features = feature_names[selected_mask]
print("\nWybrane cechy (wariancja > 0.01):")
for feat in selected_features:
    print("-", feat)

print("\nNowy kształt danych:", X_reduced.shape)

# Opcjonalnie: zapisz zredukowany zbiór
np.save("X_variance_reduced.npy", X_reduced)
```
