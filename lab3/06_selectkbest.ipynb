```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import StandardScaler

# Wczytaj dane
data = load_breast_cancer()
X, y = data.data, data.target
feature_names = data.feature_names

# Standaryzacja (f_classif działa też bez tego, ale warto dla spójności)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# SelectKBest – wybierz 10 najlepszych cech
k = 10
selector = SelectKBest(score_func=f_classif, k=k)
X_selected = selector.fit_transform(X_scaled, y)

# Wybierz nazwy wybranych cech
mask = selector.get_support()
selected_features = feature_names[mask]

# Zapisz dane
np.save("X_selectkbest.npy", X_selected)

df_selected = pd.DataFrame(X_selected, columns=selected_features)
df_selected['target'] = y
df_selected.to_csv("X_selectkbest.csv", index=False)

with open("selected_features_kbest.txt", "w") as f:
    f.write(f"Wybrane cechy (SelectKBest, k={k}):\n")
    for feat in selected_features:
        f.write(f"- {feat}\n")

print("Zapisano pliki:")
print("- X_selectkbest.npy")
print("- X_selectkbest.csv")
print("- selected_features_kbest.txt")
```
