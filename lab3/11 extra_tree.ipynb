```python
# Krok 1: Importy
from sklearn.datasets import load_iris
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Krok 2: Wczytanie danych
iris = load_iris()
X = iris.data
y = iris.target

# Krok 3: Podział na zbiór treningowy i testowy
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Krok 4: Trening modelu ExtraTreesClassifier
clf = ExtraTreesClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Krok 5: Predykcja
y_pred = clf.predict(X_test)

# Krok 6: Ewaluacja
print("Classification Report:\n")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:\n")
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# (Opcjonalnie) Ważność cech
import pandas as pd
feature_importances = pd.Series(clf.feature_importances_, index=iris.feature_names)
feature_importances.sort_values(ascending=True).plot(kind='barh')
plt.title("Feature Importances")
plt.show()
```
