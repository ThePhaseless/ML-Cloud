```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

# 1. Wczytaj dane
data = load_breast_cancer()
X, y = data.data, data.target
feature_names = data.feature_names
target_names = data.target_names

# 2. Zamień dane do DataFrame
df = pd.DataFrame(X, columns=feature_names)
df['target'] = y

# 3. Ustawienia wykresu
n_cols = 5
n_rows = int(np.ceil(len(feature_names) / n_cols))

plt.figure(figsize=(20, n_rows * 3))

# 4. Rysuj histogramy dla każdej cechy
for i, feature in enumerate(feature_names):
    plt.subplot(n_rows, n_cols, i + 1)
    for cls in np.unique(y):
        plt.hist(df[df['target'] == cls][feature],
                 bins=20, alpha=0.5, label=target_names[cls])
    plt.title(feature)
    plt.xticks([])
    plt.yticks([])
    plt.legend()

plt.suptitle("Histogramy cech – naoczna analiza istotności", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
