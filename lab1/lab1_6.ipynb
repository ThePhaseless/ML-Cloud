```python
import pandas as pd
import matplotlib.pyplot as plt

# Tworzymy przykładowe dane (jeśli plik nie istnieje)
data = {
    "Imię": ["Anna", "Jan", "Kasia", "Tomek", "Ola"],
    "Wiek": [25, 30, 22, 28, 35],
    "Miasto": ["Warszawa", "Kraków", "Gdańsk", "Wrocław", "Poznań"],
    "Zarobki": [4500, 5200, 3800, 5000, 6200]
}

# Tworzymy DataFrame
df = pd.DataFrame(data)

# Zapisujemy dane do pliku CSV
df.to_csv("dane.csv", index=False)
print("Plik dane.csv został zapisany!")

# Wczytujemy plik CSV do DataFrame
df = pd.read_csv("dane.csv")

# Wyświetlamy pierwsze 5 wierszy
print("Dane wczytane z pliku CSV:")
print(df.head())

# Obliczamy średni wiek
sredni_wiek = df["Wiek"].mean()
print("\nŚredni wiek:", sredni_wiek)

# Filtrujemy osoby zarabiające powyżej 5000 zł
wysokie_zarobki = df[df["Zarobki"] > 5000]
print("\nOsoby zarabiające powyżej 5000 zł:")
print(wysokie_zarobki)

# Dodajemy nową kolumnę - kategoria wiekowa
df["Kategoria Wiekowa"] = df["Wiek"].apply(lambda x: "Młody" if x < 30 else "Dorosły")

# Sortowanie danych według wieku malejąco
df_sorted = df.sort_values(by="Wiek", ascending=False)
print("\nDane posortowane według wieku malejąco:")
print(df_sorted)

# Grupowanie danych według kategorii wiekowej i obliczanie średnich zarobków
srednie_zarobki = df.groupby("Kategoria Wiekowa")["Zarobki"].mean()
print("\nŚrednie zarobki w każdej kategorii wiekowej:")
print(srednie_zarobki)

# Tworzymy wykres słupkowy zarobków
plt.figure(figsize=(8, 5))
plt.bar(df["Imię"], df["Zarobki"], color="blue")

# Dodajemy tytuł i etykiety
plt.title("Zarobki osób", fontsize=14)
plt.xlabel("Imię", fontsize=12)
plt.ylabel("Zarobki (PLN)", fontsize=12)
plt.xticks(rotation=45)  # Obracamy nazwy na osi X

# Wyświetlamy wykres
plt.show()
```
