```python
import pandas as pd

# Tworzymy bardziej rozbudowany DataFrame
data = {
    "Imię": ["Anna", "Jan", "Kasia", "Tomek", "Ola"],
    "Wiek": [25, 30, 22, 28, 35],
    "Miasto": ["Warszawa", "Kraków", "Gdańsk", "Wrocław", "Poznań"],
    "Zarobki": [4500, 5200, 3800, 5000, 6200]
}

df = pd.DataFrame(data)

# Wyświetlamy dane
print("Pełna tabela:")
print(df)

# **1. Obliczamy średni wiek**
sredni_wiek = df["Wiek"].mean()
print("\nŚredni wiek:", sredni_wiek)

# **2. Filtrujemy osoby zarabiające powyżej 5000 zł**
wysokie_zarobki = df[df["Zarobki"] > 5000]
print("\nOsoby zarabiające powyżej 5000 zł:")
print(wysokie_zarobki)

# **3. Dodajemy nową kolumnę - kategoria wiekowa**
df["Kategoria Wiekowa"] = df["Wiek"].apply(lambda x: "Młody" if x < 30 else "Dorosły")

# **4. Wyświetlamy zaktualizowany DataFrame**
print("\nTabela z kategoriami wiekowymi:")
print(df)
```
